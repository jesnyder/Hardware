```python
import matplotlib.pyplot as plt;  import numpy as np; import csv; import os;
```

```python
def radiation_clones(filename, title, plot_number):
    
    ax = fig.add_subplot(plot_number);
    
    file = os.path.join("Data", filename + ".csv")
    droplets_printed = 400; 
    radiation = []; clones = []; surviving_fraction = []; data = [];

    f = open(file,'r')
    for line in f: data.append(line)
    f.close() 

    for i in range(len(data)):
        x = data[i].split(",")
        if i>0:
            radiation.append(int(x[0]))
            clones.append(int(x[1]))

    plt.scatter(radiation, clones, color = "none", edgecolors = [0,0,1])
    plt.title(title)
    plt.xlabel('X-Ray Radiation (Gy)'); 
    plt.ylabel('Colonies'); 

    
    
    return(radiation, clones)  
```

```python
files = ["A549 5", "MCF7 6", "A549 9" ];
titles = ["A549 4 Days after Radiation", "MCF7 6 Days after Radiation", "A549 9 Days after Radiation"];


box_size = 3; fig = plt.figure(1, figsize=(4*box_size, 1.25*box_size)); 
plot_number = 131;

for i in range(len(files)):
    a,b = radiation_clones(files[i], titles[i], plot_number+i)

plt.show()


```
