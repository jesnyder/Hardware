```python
import matplotlib.pyplot as plt
import numpy as np

# Function to parse data into time, voltage, and impedence data
def parse_data(filename):
    data = []; time = []; voltage = []; impedence = [];

    f = open(filename,'r')
    for line in f: data.append(line)
    f.close()
    
    for i in range(len(data)):
        t, v, i = data[i].split(",")
        time.append(float(t))
        voltage.append(float(v))
        impedence.append(float(i))
        
    return (time, voltage, impedence)

def plot_data(specimen_name):
    
    time, voltage, impedence = parse_data(specimen_name + "AfterProbe_Cyclic.txt")
    plt.plot(voltage, impedence)

    time, voltage, impedence = parse_data(specimen_name + "AfterTarget_Cyclic.txt")
    plt.plot(voltage, impedence)
    
    plt.xlabel('Voltage (V)')
    plt.ylabel('Impednece (Amps))')
    
    plt.show() 
    
    

specimen_name = "1_B_S_P_"; plot_data( specimen_name )
specimen_name = "2_B_S_P_"; plot_data( specimen_name )
specimen_name = "3_N_S_P_"; plot_data( specimen_name )
specimen_name = "4_N_S_P_"; plot_data( specimen_name )
specimen_name = "5_N_N_P_"; plot_data( specimen_name )
specimen_name = "6_N_N_N_"; plot_data( specimen_name )
specimen_name = "7_B_N_N_"; plot_data( specimen_name )
```
