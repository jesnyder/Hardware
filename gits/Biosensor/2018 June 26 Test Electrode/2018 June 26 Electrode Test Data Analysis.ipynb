```python
import matplotlib.pyplot as plt; import numpy as np; import os

# Function to parse data into time, voltage, and impedence data
def parse_data(filename):
    data = []; time = []; voltage = []; impedence = [];
    f = open(filename,'r')
    for line in f: data.append(line)
    f.close()
    
    for i in range(len(data)):
        t, v, i = data[i].split(",")
        time.append(float(t))
        voltage.append(float(v))
        impedence.append(float(i))
        
    return (time, voltage, impedence)

def plot_data(specimen_name_list):    
    measurement_types = ["CyclicVoltage", "SquareWaveVoltage"] 
    measurement_name = ["Cyclic", "Square"]
    measurement_take = ["AfterProbe_", "AfterTarget_"]
    plot_color = ["red","blue","yellow","black","green","pink","purple"]
    a = [.8, .8, .8]
    b = [.8, .8, .8]
    c = [.8, .8, .8]
    d = [1,0,0]
    e = [0,0,1]
    f = [.2,.2,.2]
    g = [.8, .8, .8]
    plot_color = [a,b,c,d,e,f,g]
    plot_linestyle = ["-","--"]
        
    for j in range(2):         
        if j == 0:
            y_max = 900; y_min = -1100;
        
        else: y_max = 225; y_min = 0;
            
        for k in range(len(specimen_name_list)): 
            specimen_name = specimen_name_list[k]
            
            for i in range(2): 
                filename = (specimen_name + measurement_take[i] + measurement_name[j] + ".txt")
                filename = os.path.join("Raw_Data", measurement_types[j], filename)
                time, voltage, impedence = parse_data(filename)
                plt.plot(voltage, impedence, color= plot_color[k], linestyle = plot_linestyle[i])
  
            plt.xlabel('Voltage (V)'); 
            plt.ylabel('Current (Amps)'); plt.ylim((y_min,y_max))
            plt.title(specimen_name + measurement_name[j])
        
            filename = os.path.join("Figures", "All Samples One Plot", measurement_name[j], specimen_name + '.jpg')
            plt.savefig(filename); 
        plt.show();  
    
specimen_name_list = ["1_B_S_P_", "2_B_S_P_", "3_N_S_P_","4_N_S_P_","5_N_N_P_", "6_N_N_N_","7_B_N_N_"]
plot_data( specimen_name_list )
```
