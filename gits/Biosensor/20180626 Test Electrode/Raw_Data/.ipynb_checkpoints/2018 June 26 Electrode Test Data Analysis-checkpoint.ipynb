```python
import matplotlib.pyplot as plt; import numpy as np; import os

# Function to parse data into time, voltage, and impedence data
def parse_data(filename):
    data = []; time = []; voltage = []; impedence = [];
    f = open(filename,'r')
    for line in f: data.append(line)
    f.close()
    
    for i in range(len(data)):
        t, v, i = data[i].split(",")
        time.append(float(t))
        voltage.append(float(v))
        impedence.append(float(i))
        
    return (time, voltage, impedence)

def plot_data(specimen_name):
    
    print(specimen_name)
    
    measurement_types = ["CyclicVoltage", "SquareWaveVoltage"] 
    measurement_name = ["Cyclic", "Square"]
    measurement_take = ["AfterProbe_", "AfterTarget_"]
    plot_color = ["red","blue"]
    plot_linestyle = [":","--"]

    for j in range(2): 
        
        for i in range(2): 
            filename = os.path.join(measurement_types[j], specimen_name + measurement_take[i] + measurement_name[j] + ".txt")
            time, voltage, impedence = parse_data(filename)
            plt.plot(voltage, impedence, color= plot_color[i], linestyle = plot_linestyle[j])
  
        plt.xlabel('Voltage (V)'); 
        plt.ylabel('Current (Amps))'); #plt.ylim((0,225))
        plt.savefig(specimen_name + measurement_name[j] + '.jpg'); 
        plt.show();  
        
specimen_name = "1_B_S_P_"; plot_data( specimen_name )
specimen_name = "2_B_S_P_"; plot_data( specimen_name )
specimen_name = "3_N_S_P_"; plot_data( specimen_name )
specimen_name = "4_N_S_P_"; plot_data( specimen_name )
specimen_name = "5_N_N_P_"; plot_data( specimen_name )
specimen_name = "6_N_N_N_"; plot_data( specimen_name )
specimen_name = "7_B_N_N_"; plot_data( specimen_name )



```
