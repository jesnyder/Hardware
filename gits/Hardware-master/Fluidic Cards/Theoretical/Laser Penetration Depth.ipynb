```python
## Calculate the penetration depth of a laser 
## Use the Beer-Lambert Law
## Defined in 1729 by Pierre Bouger, modified in 1852 by August Beer  
## The absorbance 
## https://en.wikipedia.org/wiki/Beer%E2%80%93Lambert_law
```

```python
## Libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Build data
wavelength = np.linspace(0,1000,50);
penetration_depth = [];
refraction_index = 1.3; # https://en.wikipedia.org/wiki/List_of_refractive_indices

for i in range(len(wavelength)):
    depth = wavelength[i]*1/(4*22/7)/refraction_index
    penetration_depth.append(depth)
    
# Plot
plt.scatter(wavelength, penetration_depth)
plt.title('Penetration Depth in Water')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Attenuation Depth (nm)')
plt.show()

# Create data
N = 500
x = np.random.rand(N)
y = np.random.rand(N)
colors = (0,0,0)
area = np.pi*3
 
# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Scatter plot pythonspot.com')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
